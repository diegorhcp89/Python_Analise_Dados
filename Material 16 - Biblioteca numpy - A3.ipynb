```python
# Import
import numpy as np
```

```python
# Exemplo 1
# Análise de crescimento percentual mês a mês

faturamento_mensal = np.array([2500, 2700, 3000, 32000, 3100, 3300, 3400, 3500, 3750, 3800, 3999, 3450])
crescimento_percentual = np.diff(faturamento_mensal) / faturamento_mensal[:-1] * 100
print(crescimento_percentual)

for i, crescimento in enumerate(crescimento_percentual, start=1):
    print(f'Do mês {i} para o mês {i + 1}: {crescimento:.2f}%')

print()
```

```python
# Exemplo 2
# Identificação dos itens de estoque fora do padrão

nomes_produtos = np.array([
    "PROD_A",  "PROD_B",  "PROD_C",  "PROD_D",  "PROD_E",  "PROD_F",
     "PROD_G",  "PROD_H",  "PROD_I",  "PROD_J",  "PROD_K",  "PROD_L"
])
estoque_produtos = np.array([100, 150, 120, 130, 160, 110, 3000, 140, 125, 135, 145, 1150])
media_estoque = np.mean(estoque_produtos)
desvio_estoque = np.std(estoque_produtos)
fora_do_padrao = np.abs(estoque_produtos - media_estoque) > 2 * desvio_estoque

# zip para iterar sobre os nomes e estoques simultaneamente
for nome, estoque in zip(nomes_produtos[fora_do_padrao], estoque_produtos[fora_do_padrao]):
    print(f'Estoque fora do padrão: {nome} - {estoque}')
```
